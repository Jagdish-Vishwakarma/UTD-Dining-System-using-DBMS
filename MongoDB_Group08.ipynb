```python
json_file_path = "UTD_Dining.Employee.json"

# Read JSON data from the file
with open(json_file_path, 'r') as json_file:
    employees_data = json.load(json_file)


# Specify the MySQL table name and column names
table_name = "Employee"
columns = ["EmployeeID", "FirstName", "LastName", "Position", "DepartmentID", "Salary"]

# Generate SQL script
sql_script = f"INSERT INTO {table_name} ({', '.join(columns)}) VALUES"

# Iterate through each employee record
for employee in employees_data:
    # Create a VALUES clause for each record
    values_clause = f"('{employee['EmployeeID']}', '{employee['FirstName']}', '{employee['LastName']}', '{employee['Position']}', '{employee['DepartmentID']}', {employee['Salary']})"
    
    # Add the VALUES clause to the SQL script
    sql_script += f"\n{values_clause},"

# Remove the trailing comma
sql_script = sql_script.rstrip(',')

# Print the generated SQL script
print(sql_script)
```

```python

# Specify the path to your JSON file
json_file_path_shifts = "UTD_Dining.Department.json"

# Read JSON data from the shifts file
with open(json_file_path_shifts, 'r') as json_file_shifts:
    department_data = json.load(json_file_shifts)


# Specify the MySQL table name and column names for shifts
table_name_shifts = "Department"
columns_shifts = ["DepartmentID", "DepartmentName"]

# Generate SQL script for shifts
sql_script_shifts = f"INSERT INTO {table_name_shifts} ({', '.join(columns_shifts)}) VALUES"

# Iterate through each shift record
for Department in department_data:
    # Create a VALUES clause for each shift record
    values_clause_shifts = f"('{Department['DepartmentID']}', '{Department['DepartmentName']}')"
    
    # Add the VALUES clause to the SQL script for shifts
    sql_script_shifts += f"\n{values_clause_shifts},"

# Remove the trailing comma
sql_script_shifts = sql_script_shifts.rstrip(',')

# Print the generated SQL script for shifts
print(sql_script_shifts)
```
